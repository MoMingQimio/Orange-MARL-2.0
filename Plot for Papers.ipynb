```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
scenario = "lattice" #博弈所处的环境，一共分为lattice,ba,random,ws四种。
n_agents = 400 #智能体的总数。
AI_rate = 1  #智能体为强化学习者的比例。

N_GAMES = 1  #一局演化重复的次数。
MAX_STEPS = 3000 #一次演化最多会有多少步

evaluate = True  #控制训练的开关。

init_prob = 0.5
spatial_distirbution_list = [0, 50, 99] #调整输出结果
```

```python
AI_rates=(np.arange(11))/ 10.0
```

```python
AI_rates
```

```python
results = []
for AI_rate in AI_rates:
    result = pd.read_csv('results/cooperate_counts/' + scenario + '/'+"evaluate = "+ str(evaluate) +", Agents_num = "+ str(n_agents) + (', AI-rate = %.2f' % AI_rate) + " .csv"  )
    results.append(result)
```

```python
plt.figure()
x = np.arange(30000-1)
for result ,AI_rate in zip(results[:],AI_rates[:]):
    plt.plot( x[:40],result.to_numpy().flatten()[:40] ,label= AI_rate )
plt.legend(AI_rates, loc = "lower right") 
```

```python
plt.plot(results)
```

```python
plt.figure()
x = np.arange(3000-1)

plt.boxplot( [result[-300:-1].to_numpy().flatten() for result in results ] )

```

```python
spatial_distirbutions = []
for episode_step in spatial_distirbution_list:
    spatial_distirbution = pd.read_csv('results/spatial_distribution/' + scenario + '/' 
                                            + scenario 
                                            + ", evaluate = " + str(evaluate) 
                                            + ", Agents_num = "+ str(n_agents) 
                                            + (', AI-rate = %.2f' % AI_rate) 
                                            + (", eposide= %d" % episode_step)
                                            + " .csv" ,names= range(int(np.sqrt(n_agents))) )
    spatial_distirbutions.append(spatial_distirbution)
```
